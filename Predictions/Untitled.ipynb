```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('white')
plt.xkcd();
import pandas as pd
from math import e
plt.rcParams['svg.fonttype'] = 'none'

# Df 1 for narrow
error=np.random.uniform(-40,40,500)

probability=np.random.randint(0,100,500)
B0=-38
B1=1

nar_choice=np.divide(1,1+e**-(B0+B1*probability+error))
nar_choice=nar_choice>0.5

df1=pd.DataFrame(data=np.array([probability,nar_choice]).T,
                columns=['Probability','Choice'])
df1['Choice']=df1['Choice'].astype(int)
df1['Distribution']='Low Variance'

#DF2 for wide
error=np.random.uniform(-40,40,500)

probability=np.random.randint(0,100,500)
B0=-38
B1=0.9

nar_choice=np.divide(1,1+e**-(B0+B1*probability+error))
nar_choice=nar_choice>0.5

df2=pd.DataFrame(data=np.array([probability,nar_choice]).T,
                columns=['Probability','Choice'])
df2['Choice']=df2['Choice'].astype(int)
df2['Distribution']='High Variance'

df=pd.concat([df1,df2])
df['Model']='MMEU'
f=sns.lmplot(x='Probability',y='Choice',data=df,hue='Distribution',logistic=True,
           x_jitter=3,y_jitter=.01)
f.savefig('No_violations_underAmbiguity.jpg')
f.axes[0][0].set_ylabel('Choose Ambiguous')
f.axes[0][0].set_xlabel('Probability Winning Risky Outcome')

plt.show()
df_MEU=df
#sns.lmplot(x='probability',y='choice',data=df2,hue='distribution')#,logistic=True)
```

# Showing the Mediation

```python
# Import libraries
#from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(42);

def make_data():
    N = 100
    a, b, cprime = -0.6, 0.6, 0.0
    im, iy, σm, σy = 10.0, 10.0, 0.5, 0.5
    x1 = rng.normal(loc=10, scale=2, size=50)
    x2 = rng.normal(loc=15, scale=2, size=50)
    x=np.concatenate([x1,x2])
    m = im + rng.normal(loc=a * x, scale=σm, size=N)
    y = iy + (cprime * x) + rng.normal(loc=b * m, scale=σy, size=N)
    print(f"True direct effect = {cprime}")
    print(f"True indirect effect = {a*b}")
    print(f"True total effect = {cprime+a*b}")
    return x, m, y


x, m, y = make_data()
clin_label=['Healthy Control']*50+['Autism Spectrum Diagnosis']*50

df=pd.DataFrame({"Autism Quotient": x, "Belief Variance": m, "Ambiguity Aversion": y,"Clinical Label":clin_label})

f=sns.lmplot(x='Belief Variance',y='Ambiguity Aversion',data=df,hue='Clinical Label');
f.savefig('ClinicalCorrelation.svg')
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
#ax.scatter3D(x, m, y, color = "green")
#plt.title("simple 3D scatter plot")
 
# show plot
plt.show()
```
